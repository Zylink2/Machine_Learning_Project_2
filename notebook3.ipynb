 > ## **Intro**
 ---
ในยุคปัจจุบัน การย้ายถิ่นฐานกลายเป็นปรากฏการณ์ทางสังคมที่พบเห็นได้ทั่วไป ผู้คนจากถิ่นกำเนิดหนึ่งเดินทางไปยังอีกถิ่นกำเนิดหนึ่งด้วยเหตุผลที่หลากหลาย เช่น การศึกษา การหางาน การแต่งงาน หรือเพื่อคุณภาพชีวิตที่ดีขึ้น
ในประเทศไทย การย้ายถิ่นฐานมีบทบาทสำคัญต่อการพัฒนาเศรษฐกิจและสังคม ผู้คนจากต่างจังหวัดย้ายเข้ามาอยู่ในเขตเมืองเพื่อหางานและโอกาสที่ดีกว่า ส่งผลให้เกิดการเปลี่ยนแปลงทางสังคม หรือแม้กระทั่งสังคมผู้สูงอายุ  (aging society)

วัตถุประสงค์ของการวิจัยครั้งนี้ คือ การเข้าไปสำรวจดูถึงสถานการณ์ aging society กับ พฤติกรรมการย้ายถิ่นฐานซึ่งสนใจกลุ่มอายุ 20-30 ปีซึ่งจะกลายเป็นกำลังหลักของครอบครัวในอนาคต

> ## **Definition**
---

*  การย้ายถิ่น หมายถึง การที่ย้ายสถานที่อยู่อาศัยจากเขตเทศบาลอื่นหรือ อบต.อื่น หรือจาก
ประเทศอื่น ซึ่งเป็นที่อยู่อาศัยครั้งสุดท้ายมายังเขตเทศบาลหรือ อบต. ซึ่งเป็นพื้นที่ที่อยู่ปัจจุบัน
การเปลี่ยนเขตเทศบาลหรือ อบต. เนื่องจาก การขยายเขตเทศบาลหรือแยกหมู่บ้าน โดยที่บุคคล
หรือครัวเรือนนั้นยังอยู่ประจำที่เดิม และการย้ายสถานที่อยู่อาศัยในกรุงเทพมหานคร และเมืองพัทยา
ไม่นับว่าเป็นการย้ายถิ่น

* ประชากรแฝง หมายถึง ประชากรที่เข้ามาอยู่อาศัย มาเรียน หรือมาทำงานในพื้นที่หนึ่ง แต่ไม่ได้ย้ายทะเบียนบ้านมาอยู่จริงในพื้นที่นั้น แบ่งออกได้เป็น 2 ประเภทหลัก ดังนี้

    1. ประชากรแฝงกลางวัน (Commuter Population)
        * หมายถึง ผู้ที่มีอายุตั้งแต่ 15 ปีขึ้นไป ที่เข้ามาทำงานหรือเรียนในพื้นที่หนึ่ง โดยไม่ได้ย้ายทะเบียนบ้านมาอยู่จริงในพื้นที่นั้น
        * กลุ่มนี้มักเดินทางไปกลับระหว่างบ้านกับสถานที่ทำงานหรือสถานศึกษา
        * ตัวอย่าง: คนที่อาศัยอยู่ในจังหวัดนนทบุรี แต่ทำงานในกรุงเทพฯ
    
    2. ประชากรแฝงกลางคืน (Non-registered Population)
        * หมายถึง ผู้ที่อาศัยอยู่ประจำในพื้นที่หนึ่ง โดยไม่มีชื่ออยู่ในทะเบียนบ้านในพื้นที่นั้น
        * กลุ่มนี้มักมาจากต่างจังหวัด หรือต่างประเทศ
        * สาเหตุที่ไม่ย้ายทะเบียนบ้าน อาจจะเพื่อสิทธิประโยชน์ในภูมิลำเนาเดิม หรือต้องการหลีกเลี่ยงภาระภาษี
        * ตัวอย่าง: แรงงานข้ามชาติที่อาศัยและทำงานในกรุงเทพฯ

* aging society หมายถึง  สังคมที่มีสัดส่วนของประชากรอายุ 60 ปีขึ้นไปมากกว่า 10% ของประชากรทั้งหมด หรือมีอายุ 65 ปีขึ้นไปมากกว่า 7% ของประชากรทั้งหมด

* เมืองโตเดี่ยว หมายถึง เมืองที่มีขนาดใหญ่และมีประชากรอาศัยอยู่หนาแน่น ซึ่งมักเป็นศูนย์กลางของเศรษฐกิจ การเมือง และวัฒนธรรมของประเทศ ตัวอย่างเมืองโตเดี่ยวในประเทศไทยคือ กรุงเทพมหานคร

```python
def translate_1(val):
    if val == "กรุงเทพมหานคร":
        return "Bangkok"
    elif val == "กลาง":
        return "Central"
    elif val == "เหนือ":
        return "North"
    elif val == "ตะวันออกเฉียงเหนือ":
        return "Notheast"
    else:
        return "South"
def translate_2(val):
    if val == "หางานทำ":
        return "Looking for a job"
    elif val == "ต้องการเปลี่ยนงาน":
        return "Changing job"
    elif val == "ต้องการรายได้เพิ่มขึ้น":
        return "Need more revenue"
    elif val == "หน้าที่การงาน":
        return "Responsibility"
    elif val == "ศึกษาต่อ":
        return "Graduating"
    elif val == "ย้ายที่อยู่อาศัย":
        return "Moving out"
    elif val == "กลับภูมิลำเนา":
        return "Going back home"
    elif val == "ติดตามครอบครัว":
        return "Following family"
    elif val == "ช่วยธุรกิจในครอบครัว":
        return "Helping family business"
    elif val == "รักษาตัว":
        return "Recovering"
    elif val == "ขาดคนดูแล":
        return "Lacking of care person"
    else:
        return "Others"
   
```

## จำนวนผู้สูงอายุจากการสำรวจ จำแนกตามลักษณะการอยู่อาศัย กลุ่มอายุ เพศ

```python
elders = pd.read_csv("https://catalogapi.nso.go.th/api/index?table=OS_01_0045_15&format=csv", encoding='utf-8')
elders.head(20)
```

```python
elders = elders[elders["year"] != 2545].copy()
elders
```

```python
def translate_3(val):
    if val == "รวม":
        return "All"
    elif val == "ชาย":
        return "Male"
    elif val == "หญิง":
        return "Female"
    else:
        return "Others"
def translate_4(val):
    if val == "รวม":
        return "All"
    elif val == "60-69 ปี":
        return "60-69 yrs old"
    elif val == "70-79 ปี":
        return "70-79 yrs old"
    elif val == "80 ปีขึ้นไป":
        return "80+ yrs old"
    else:
        return "Others"
def translate_5(val):
    if val == "รวม":
        return "All"
    elif val == "อยู่คนเดียว":
        return "Alone"
    elif val == "อยู่หลายคน":
        return "Not alone"
    else:
        return "Others"
    
```

```python
elders[['sex', 'age_group','type_of_con_mat']] = elders[['sex', 'age_group','type_of_con_mat']].apply(lambda x: x.str.strip())
elders['sex'] = elders['sex'].apply(lambda x : translate_3(x))
elders['age_group'] = elders['age_group'].apply(lambda x : translate_4(x))
elders['type_of_con_mat'] = elders['type_of_con_mat'].apply(lambda x : translate_5(x))
elders = elders[(elders['type_of_con_mat'] == 'Alone') & (elders['age_group'] != 'All') & (elders['sex'] != 'All')].copy()
elders.drop(columns=["unit","source"], inplace=True)
elders
```

# Step 5 : Cause and Solution
---


   หลังจากที่เราได้ดูและวิเคราะห์กราฟต่างๆ จากด้านบนแลัว ต่อมาเราจะมาดูสาเหตุและวิธีแก้ไขปัญหาใน เรื่องอ้ตราเจริญพันธุ์ที่ต่ำลงที่มีและคนย้ายเข้าเมือง ที่ส่งผลใน เรื่อง aging society


> Causes

* ต้นเหตุที่ทำให้อัตราเจริญพันธุ์ต่ำลง
    * ต้นทุนชีวิตที่สูงขึ้นซึ่งไม่สอดคล้องกับค่าตอบแทนจากการทำงาน ซึ่งเป็นหนึ่งในปัจจัยที่สำคัญที่ทำให้คนรุ่นใหม่ส่วนใหญ่ตัดสินใจไม่มีลูก เพราะ เฉพาะการหาเงินเพื่อใช้เลี้ยงชีพก็ยากแล้ว 
      ถ้าเกิดมีลูกอีกก็จะส่งผลให้มีรายจ่ายที่เพิ่มขึ้นอีก
    * คนรุ่นใหม่ส่วนใหญ่ให้ความสำคัญกับอิสระและความยืดหยุ่นในการใช้ชีวิต การมีลูกหมายถึงการมีภาระหน้าที่และความรับผิดชอบเพิ่มมากขึ้น ซึ่งอาจจำกัดเวลาส่วนตัวและโอกาสในการทำสิ่งที่ชอบ
    * เมื่อมองในสถานการณ์ของประเทศไทย อีกหนึ่งปัจจัยที่ทำให้คนไม่อยากมีลูก อาจเป็นเรื่องการเมืองหรือเศรษฐกิจทีไม่มึความมั้นคงในประเทศ ซึ่งจะส่งผลกระทบต่อความเชื่อมั่นและการวางแผนอนาคตของประชาชน     
    กล่าวคือ เพราะหากเศรฐกิจและการเมืองไม่มั่นคงในอนาคต จะทำให้ครอบครัวซึ่งรวมถึงลูกมีความลำบาก ส่งผลให้คนไม่อยากมีลูก

* ต้นเหตุที่ทำให้คนต้องเข้าไปทำงานในเมืองมากขึ้น
    * ปัญหาเมืองโตเดี่ยว
        *  โอกาสในการประกอบอาชีพและรายได้ที่สูงกว่าเมืองอื่นๆ  ด้วยความเป็นศูนย์กลางทางเศรษฐกิจ การเมือง การศึกษา และวัฒนธรรม กรุงเทพฯ ดึงดูดผู้คนจากทั่วประเทศให้เข้ามาหางานและแสวงหาโอกาสที่ดีกว่
        *  หนึ่งในแรงดึงดูดหลักของกรุงเทพมหานคร เช่น ระบบคมนาคมที่สะดวกกว่าต่างจังหวัด ซึ่งก็ทำให้ชีวิตสะดวกสบายมากขึ้น  นอกจากนี้ยังเป็นเรื่องของการเข้าถึงทรัพยากรต่างๆได้อย่างสะดวก เช่น ห้างสรรพสินค้า บริการต่างๆที่รองรับเพียงในกรุงเทพอย่างบริการ food delivery เป็นต้น





> Solution
- วิธีแก้ปัญหาอัตราเจริญพันธุ์ที่ต่ำลง
    * เพิ่มสวัสดิการที่ช่วยให้คนมีลูกมากขึ้น เช่น การเพิ่มเงินช่วยเหลือค่าใช้จ่าย โดยเฉพาะช่วงวัยเด็กเล็ก, การอุดหนุนค่าศึกษาขั้นพื้นฐานฟรี และรวมไปถึงการช่วยเหลือค่าใช้จ่ายด้านอสังหาริมทรัพย์ เพื่อสนับสนุนให้คนรุ่นใหม่มีบ้านในราคาที่เข้าถึงได้ เหมาะกับการสร้างครอบครัว
    * เพิ่มความเชื่อมั่นทางการเมือง แม้จะดูเป็นเรื่องเกินจะควบคุมได้ แต่รัฐบาลควรแสดงท่าทีและออกนโยบายเพื่อเพิ่มเสถียรภาพทางการเมืองมากขึ้น รวมไปการลดความขัดแย้งการทางการเมือง เพื่อทำให้คนรุ่นใหม่เชื่อว่า  ภาพในอนาคตเหมาะสมกับการสร้างครอบครัว


- วิธีแก้ปัญหาคนย้ายเข้าเมือง 
    * กระจายอำนาจการปกครองสู่ท้องถิ่น โดยการให้คนในท้องถิ่นมีอำนาจในการเลือกคนที่จะบริหารการปกครองส่วนท้องถิ่นของตัวเอง เพื่อสะท้อนความต้องการของชุมชนอย่างแท้จริง
    * เพิ่มผลตอบแทนเพื่อดึงดูดให้คนที่มีความสามารถกลับเข้าไปทำงานที่บ้านเกิดเพื่อพัฒนาบ้านเกิดของตัวเอง ซึ่งอาจจะเป็นการลดภาษี หรือลดดอกเบี้ยเงินกู้ให้กับบริษัทที่เปิด headquaters ที่ต่างจังหวัด เป็นต้น
    * จัดสรรงบประมาณเพิ่มให้กับท้องถิ่น เพื่อให้ท้องถิ่นสามารถควบคุมการเก็บภาษี หรือจัดสรรงบประมาณได้อย่างมีอิสระมากขึ้น รวมไปถึงการจัดให้มีสาธารณูปโภคที่จำเป็นต่อการเติบโต เช่น ระบบขนส่งสาธารณะ เป็นต้น

จากกราฟต่างๆ รวมถึงปัญหาและสาเหตุที่กลุ่มเราได้วิเคราะห์และค้นพบ จึงทำให้สรุปได้ว่า ผลกระทบที่สำคัญของการย้ายถิ่นฐานที่มีต่อ aging society คือ ทำให้ผู้สูงอายุถูกทิ้งให้อยู่คนเดียวมากขึ้น ซึ่งสาเหตุที่ทำให้คนย้ายถิ่นฐานมีทั้งเรื่อง การหางาน การศึกษา หรือแม้กระทั่งการใช้ชีวิตที่มีความสะดวกสะบายกว่าต่างจังหวัด
ไม่ว่าจะเป็นการย้ายถิ่นฐานแบบย้ายทะเบียนและไม่ย้ายทะเบียนบ้าน และนอกจากสาเหตุการย้ายถิ่นฐานแลัว ยังมีสาเหตุหลักที่อีกประการหนึ่งที่ทำให้ผู้สูงอายุถูกทิ้งไว้คนเดียว
คือ เรื่องอัตราเจริญพันธุ์ที่ต่ำลง ที่เกิดขึ้นมาจาก การที่ต้นทุนชีวตคนสูงขึ้น คนส่วนมากชอบใช้ที่ชีวิตยืดหยุ่น หรือ สภาพเศรษฐกิจในประเทศ เป็นต้น ด้วยสาเหตุทั้งหมดนี้ จึงทำให้ต้องมีวิธีการรับมือที่เหมาะกับเหตุการณ์ที่เกิดใน aging society
