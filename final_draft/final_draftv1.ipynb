### ข้อมูลอัตราการดื่มสุราของประชากรอายุ 15-24 ปี

```python
import matplotlib.pyplot as plt

alcoholic = pd.read_csv("https://apis1.nso.go.th/data?table=OS_05_0006_01&format=csv")
alcoholic
```

## Analysis
---

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.barplot(data = alcoholic, x = 'year', y = 'value', color='gray', ax=axes[0]);
sns.regplot(x='year', y='value', data=alcoholic, scatter=True, ci=None, color='gray', line_kws={"color":"tomato","alpha":0.7,"lw":2}, ax = axes[1]);
```
