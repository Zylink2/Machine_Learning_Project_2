 > ## **Intro**
 ---
ในยุคปัจจุบัน การย้ายถิ่นฐานกลายเป็นปรากฏการณ์ทางสังคมที่พบเห็นได้ทั่วไป ผู้คนจากถิ่นกำเนิดหนึ่งเดินทางไปยังอีกถิ่นกำเนิดหนึ่งด้วยเหตุผลที่หลากหลาย เช่น การศึกษา การหางาน การแต่งงาน หรือเพื่อคุณภาพชีวิตที่ดีขึ้น
ในประเทศไทย การย้ายถิ่นฐานมีบทบาทสำคัญต่อการพัฒนาเศรษฐกิจและสังคม ผู้คนจากต่างจังหวัดย้ายเข้ามาอยู่ในเขตเมืองเพื่อหางานและโอกาสที่ดีกว่า ส่งผลให้เกิดการเปลี่ยนแปลงทางสังคม วัฒนธรรม และสิ่งแวดล้อม








> ## **Definition**
---

*  การย้ายถิ่น หมายถึง การที่ย้ายสถานที่อยู่อาศัยจากเขตเทศบาลอื่นหรือ อบต.อื่น หรือจาก
ประเทศอื่น ซึ่งเป็นที่อยู่อาศัยครั้งสุดท้ายมายังเขตเทศบาลหรือ อบต. ซึ่งเป็นพื้นที่ที่อยู่ปัจจุบัน
การเปลี่ยนเขตเทศบาลหรือ อบต. เนื่องจาก การขยายเขตเทศบาลหรือแยกหมู่บ้าน โดยที่บุคคล
หรือครัวเรือนนั้นยังอยู่ประจำที่เดิม และการย้ายสถานที่อยู่อาศัยในกรุงเทพมหานคร และเมืองพัทยา
ไม่นับว่าเป็นการย้ายถิ่น

* ประชากรแฝง หมายถึง ประชากรที่เข้ามาอยู่อาศัย มาเรียน หรือมาทำงานในพื้นที่หนึ่ง แต่ไม่ได้ย้ายทะเบียนบ้านมาอยู่จริงในพื้นที่นั้น แบ่งออกได้เป็น 2 ประเภทหลัก ดังนี้

    1. ประชากรแฝงกลางวัน (Commuter Population)
        * หมายถึง ผู้ที่มีอายุตั้งแต่ 15 ปีขึ้นไป ที่เข้ามาทำงานหรือเรียนในพื้นที่หนึ่ง โดยไม่ได้ย้ายทะเบียนบ้านมาอยู่จริงในพื้นที่นั้น
        * กลุ่มนี้มักเดินทางไปกลับระหว่างบ้านกับสถานที่ทำงานหรือสถานศึกษา
        * ตัวอย่าง: คนที่อาศัยอยู่ในจังหวัดนนทบุรี แต่ทำงานในกรุงเทพฯ
    
    2. ประชากรแฝงกลางคืน (Non-registered Population)
        * หมายถึง ผู้ที่อาศัยอยู่ประจำในพื้นที่หนึ่ง โดยไม่มีชื่ออยู่ในทะเบียนบ้านในพื้นที่นั้น
        * กลุ่มนี้มักมาจากต่างจังหวัด หรือต่างประเทศ
        * สาเหตุที่ไม่ย้ายทะเบียนบ้าน อาจจะเพื่อสิทธิประโยชน์ในภูมิลำเนาเดิม หรือต้องการหลีกเลี่ยงภาระภาษี
        * ตัวอย่าง: แรงงานข้ามชาติที่อาศัยและทำงานในกรุงเทพฯ

## - ข้อมูลเหตุผลที่ย้ายถิ่นฐาน

```python
reasons_63 = pd.read_csv('https://github.com/Zylink2/Machine_Learning_Project_2/raw/main/files/%E0%B9%80%E0%B8%AB%E0%B8%95%E0%B8%B8%E0%B8%9C%E0%B8%A5%E0%B8%82%E0%B8%AD%E0%B8%87%E0%B8%81%E0%B8%B2%E0%B8%A3%E0%B8%A2%E0%B9%89%E0%B8%B2%E0%B8%A2%E0%B8%96%E0%B8%B4%E0%B9%88%E0%B8%99_63.csv')
reasons_63["Year"] = "2563"
reasons_63[["กรุงเทพมหานคร","กลาง","เหนือ","ตะวันออกเฉียงเหนือ","ใต้"]] = reasons_63[["กรุงเทพมหานคร","กลาง","เหนือ","ตะวันออกเฉียงเหนือ","ใต้"]].round()
reasons_63
```

```python
reasons_63.fillna(reasons_63["กรุงเทพมหานคร"].quantile(0.25), inplace=True)
reasons_63
```

```python
reasons_64 = pd.read_csv('https://github.com/Zylink2/Machine_Learning_Project_2/raw/main/files/%E0%B9%80%E0%B8%AB%E0%B8%95%E0%B8%B8%E0%B8%9C%E0%B8%A5%E0%B8%82%E0%B8%AD%E0%B8%87%E0%B8%81%E0%B8%B2%E0%B8%A3%E0%B8%A2%E0%B9%89%E0%B8%B2%E0%B8%A2%E0%B8%96%E0%B8%B4%E0%B9%88%E0%B8%99_64.csv')
reasons_64.rename(columns={"เหตุผลของการย้ายถิ่น 2564":"เหตุผลของการย้ายถิ่น"}, inplace=True)
reasons_64["Year"] = "2564"
reasons_64[["กรุงเทพมหานคร","กลาง","เหนือ","ตะวันออกเฉียงเหนือ","ใต้"]] = reasons_64[["กรุงเทพมหานคร","กลาง","เหนือ","ตะวันออกเฉียงเหนือ","ใต้"]].apply(lambda x : x.str.strip().str.replace(",","")).astype(float).round()
reasons_64
```

```python
reasons_64.isnull()
```

```python
reasons_64.describe()
```

```python
reasons_64.fillna(reasons_64["กรุงเทพมหานคร"].quantile(0.25), inplace=True)
reasons_64
```

```python
reasons_6364 = pd.concat([reasons_63, reasons_64], ignore_index=True)
reasons_6364
```

```python
def acronymn(val):
    if val == "กรุงเทพมหานคร":
        return "Central"
    elif val == "กลาง":
        return "Middle"
    elif val == "เหนือ":
        return "North"
    elif val == "ตะวันออกเฉียงเหนือ":
        return "Notheast"
    else:
        return "South"
```

```python
reasons_6364 = pd.melt(reasons_6364,id_vars=["เหตุผลของการย้ายถิ่น","Year"],var_name="Region", value_name="People").copy()
reasons_6364
```

```python
reasons_6364["Region"] = reasons_6364["Region"].apply(lambda x: acronymn(x))
reasons_6364
```

```python
bargraph = sns.barplot(data=reasons_6364, x="Year", y="People", errorbar=None, hue="Region", palette="Reds_d")
bargraph.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f"{x/1000:,.1f}m"))
plt.title("Increase of Migration within Thailand across all regions between 2563 and 2564")
```

```python
fig, ax = plt.subplots(1,2, figsize=(14,12), sharey=True)
grid1 = sns.barplot(elders, x = 'year', hue = 'age_group', y = 'value', palette='Blues_d', errorbar=None, ax=ax[0])
grid2 = sns.boxplot(elders, x = 'year', hue = 'age_group', y = 'value', palette='Blues_d', ax=ax[1])
```

```python
g = sns.catplot(
    data=elders,
    kind="bar",
    y="age_group",
    x="value",
    col="year",
    row="age_group",
    color=".7",
    height=3,
    aspect=1.6,
    sharex=True,
    errorbar=None
)

def millions_formatter(x, pos):
    return f"{x/1000:.0f}K"

for ax in g.axes.flat:
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(millions_formatter))
```
